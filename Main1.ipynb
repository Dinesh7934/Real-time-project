```python
import pandas as pd
import numpy as np
import sklearn 
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Final.csv")
df
```

```python
df = df.drop(df.columns[0],axis=1)
df
```

```python
df["rating"].unique()
```

```python
df["genre"].unique()
```

```python
df["country"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
leR = LabelEncoder()
```

```python
leR.fit(df["rating"])
```

```python
list(leR.classes_)
```

```python
arr = leR.transform(df["rating"])
```

```python
rating = pd.DataFrame(arr,columns=['rating'])
rating
```

```python
df = df.drop(df.columns[0],axis=1)
df
```

```python
df=pd.concat([df,rating], axis =1)
df
```

```python
leG = LabelEncoder()
```

```python
leG.fit(df["genre"])
list(leG.classes_)
```

```python
arr = leG.transform(df["genre"])
```

```python
genre = pd.DataFrame(arr,columns=['genre'])
genre
```

```python
df = df.drop(df.columns[0],axis=1)
df
```

```python
df=pd.concat([df,genre], axis =1)
df
```

```python
leC = LabelEncoder()
```

```python
leC.fit(df["country"])
list(leC.classes_)
```

```python
arr = leC.transform(df["country"])
```

```python
country = pd.DataFrame(arr,columns=['country'])
country
```

```python
df = df.drop(df.columns[0],axis=1)
df
```

```python
df=pd.concat([df,country], axis =1)
df
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
```

```python
X = df.drop(df.columns[[1]],axis=1)
Y = df.gross
```

```python
X.head
```

```python
np.random.seed(42)
X_train ,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
model.fit(X_train,Y_train)
print("Trained!")
```

```python
model.score(X_test,Y_test)
```
